<b> <span style="font-size: 200%"> <span style='font-family:Helvetica'> The Main Dataset</b> 
* When we downloaded our data from the the Hellinikon weather station we choosed Metric option in Unit field in order to have the appropriate units of measurement
* Next we proceed by reading the Hellinikon weather station csv file for the dates 1955-2020
* We sort our data to be sure they follow chronological order based on column Date

```python
# importing/reading the dataset "hell.csv" dataset with "read_csv" method
df_hell = pd.read_csv("hell.csv", parse_dates=['DATE'])
df_hell = df_hell.sort_values(by='DATE', ignore_index=True)
df_hell
```

* Our dataset contains 23251 rows and 7 columns (variables)
* At first glance we can see that some columns e.g. "Station", "Name" they do not offer any important information and we can exlude them
* We can see some null values (in TAVG column for example)
* Also we can run the "describe" command to have a look in the statistical summary of the dataframe

```python
# descibe method  used for calculating some statistical data like percentile, mean and std of the numerical values
df_hell.describe()
```

* From the output we can focus on the mean of its variable and the mix max, to have a straight forward view about the range of its variable
* Next we proceed by checking for the total missing values in our dataset per column

```python
# checking the summary of missing values per column in "hell.csv" dataset
df_hell.isnull().sum()
```

* We have a huge amount of empty values in PRCP, TAVG, TMAX and TMIN column
* Also we can check the each column type

```python
df_hell.dtypes
```

* Next we proceed by exluding the unwanted columns Station and Name

```python
# With loc method we select all rows and we exlude (~) the two columns of dataframe
df_hell = df_hell.loc[:, ~df_hell.columns.isin(['STATION', 'NAME'])]
df_hell
```

<b> <span style="font-size: 200%"> <span style='font-family:Helvetica'> The Supplementary Dataset</b> 

* As we saw before our main dataset contain many empty data.
* For that reason we will import the "athens.csv" dataset in order fullfil the empty values of "hell.csv" dataset
* We use the same method "read_csv" to read our data
* We renamed the columns according to the description of the data in the website
* The rain column is the daily rainfall (mm)

```python
# importing the athens.csv dataset
df_ath = pd.read_csv('athens.csv', header=None)
df_ath = df_ath[[0, 1, 2, 3, 10]]
df_ath.rename(columns = {0:'DATE', 1:'TAVG', 2:'TMAX', 3:'TMIN', 10:'RAIN' }, inplace = True)
df_ath = df_ath.sort_values(by='DATE', ignore_index=True)
df_ath
```

* Next we proceed with checking the column data types

```python
df_ath.dtypes
```

* The the types can help us decide what conversion we must make later
* Next we remove the special characters from the TAVG column in order to convert into the appropriate data type



```python
#Finding and removing the characters "---" from the TAVG column
df_ath.loc[df_ath['TAVG'].str.contains("---", case=False)]
df_ath['TAVG'].replace({'---': None},inplace =True)
df_ath.loc[df_ath['TMAX'].str.contains("---", case=False)]
df_ath['TMAX'].replace({'---': None},inplace =True)
df_ath.loc[df_ath['TMIN'].str.contains("---", case=False)]
df_ath['TMIN'].replace({'---': None},inplace =True)
```

* Next we proceed with converting the columns TAVG and Date to the appropriate data types

```python
#Converting to the appropriate data types
df_ath['TAVG'] = pd.to_numeric(df_ath['TAVG'])
df_ath['TMAX'] = pd.to_numeric(df_ath['TMAX'])
df_ath['TMIN'] = pd.to_numeric(df_ath['TMIN'])
df_ath['DATE'] = pd.to_datetime(df_ath['DATE'])
```

<b> <span style="font-size: 200%"> <span style='font-family:Helvetica'> Merging the 2 Datasets</b> 


* Now we are ready to proceed merging the two files (main and supplementary)
* We will use "merge" function offered by pandas library with outer join method using as key the column "DATE"
* Specifically, the outer join (also known as full join) will keep all rows from both data frames
* The new dataframe will be named "df_merge_after"


```python
# using merger function followed by outer join to merge the two csv files
df_merged_after = df_hell.merge(df_ath, on=['DATE'], how='outer')
df_merged_after = df_merged_after.sort_values(by='DATE')
df_merged_after
```

* Next we must fill the null values from (TAVG_y, TMAX_y,TMIN_y, RAIN) collumns to fill the (PRCP, TAVG_x, TMAX_x,TMIN_x) 

```python
#using combine_first method to fill null values with non-null values 
df_merged_after['TAVG_x'] = df_merged_after['TAVG_x'].combine_first(df_merged_after['TAVG_y'])
df_merged_after['TMAX_x'] = df_merged_after['TMAX_x'].combine_first(df_merged_after['TMAX_y'])
df_merged_after['TMIN_x'] = df_merged_after['TMIN_x'].combine_first(df_merged_after['TMIN_y'])
df_merged_after['PRCP'] = df_merged_after['PRCP'].combine_first(df_merged_after['RAIN'])
df_merged_after
```

* Now we can drop the unwanted columns 
* Rename the final columns without the "_x"

```python
df_merged_after = df_merged_after.drop(['TAVG_y','TMAX_y','TMIN_y','RAIN', 'TMAX_x','TMIN_x'], axis=1)
df_merged_after =  df_merged_after.rename(columns = {'PRCP_x':'PRCP', 'TAVG_x':'TAVG' })
df_merged_after
```

<b> <span style="font-size: 200%"> <span style='font-family:Helvetica'> <span > Filtering Summer Months for All Years</b> 
* Firstly, we have to filter only summer months (June, July, August) based on column Date for all years
* So we create a new dataframe named df_all as a copy from the df_merged_after dataframe
* Then we append the "dt.month.between" command to filter all month between June and August

```python
# Creating new dataframe for summer months with dt.month.between function
df_all = df_merged_after[df_merged_after['DATE'].dt.month.between(6,8)].copy()
df_all
```

* Then we group the Date column by the year and we calculate the mean PRCP and TAVG
* We reset the index because the old index was per day not per year

```python
# Group Date Column by Year
df_all = df_all.groupby(df_all['DATE'].dt.year).mean().reset_index() 
df_all
```

<b> <span style="font-size: 200%"> <span style='font-family:Helvetica'> Filtering Summer Months for Years 1974-1999</b> 

* Next we proceed by creating a new dataframe for the requested years 1974-1999 & summer months

```python
#Creating the new dataframe for the required years 1974-1999
df_74 = df_merged_after.loc[(df_merged_after['DATE'] >= '1974') & (df_merged_after['DATE'] <= '2000') & 
                        (df_merged_after['DATE'].dt.month.between(6,8))]
df_74
```

* We reset the index because the old index was per day not per year

```python
#group by years
df_74 = df_74.groupby(df_74['DATE'].dt.year).mean().reset_index() 
df_74.head(10)
```

<b> <span style="font-size: 200%"> <span style='font-family:Helvetica'>Calculating our Baseline Mean</b> 

* Next we proceed by calculating the mean temperature for the requested years 1974-1999

```python
# calculating the mean of TAVG column with appending ".mean()" command
baseline = df_74.TAVG.mean()
baseline
```

* We proceed by adding a new collumn in our dataset named DIFF 
* The DIFF column contains the deviation of every record from the mean temperature

```python
# calculating the difference/deviation from every year average temperature (TAVG column) with the baseline's
df_all['DIFF'] = df_all['TAVG'] - baseline
```

* Next we calculate a 10 year rolling average from the mean with minimum number of observations
* Therefore we will add a new column named "ROLL"

```python
# creating rolling average
df_all['ROLL'] = df_all['DIFF'].rolling(10, min_periods=1).mean()
```

<b> <span style="font-size: 200%"> <span style='font-family:Helvetica'>Let's Start Plotting</b> 

```python
##Creating the plot
plt.figure(figsize=(14,10))
ax = plt.subplot()
plt.plot(df_all['DATE'], df_all['ROLL'],color='#4E92F7' )
colors = df_all.DIFF
f = lambda x: 'orange' if x > 0 else '#7377F9'
colors = colors.map(f)
ax.bar(df_all['DATE'], df_all['DIFF'], width=.20, color=colors)
plt.title(' Summer mean temperature (°C) relative to 1974-1999' , 
             fontdict={'fontsize': 20, 'fontweight': 'bold'})
plt.axhline(0,color='grey',xmin=0.046,xmax=0.954);
```

> * Extreme events are defined as those beyond 5% or 10% from the expected value. </br>
> * We will deal with extreme heat events going 10% above the baseline.</br>
> * We count the number of extreme temperature events per year, compared to the baseline of 1974-1999 and afterwards we will produce a a graph.</br>

<b> <span style="font-size: 200%"> <span style='font-family:Helvetica'>New Dataframe</b> 


```python
# creating a copy from the previous dataframe we used
df_q4=df_q3.copy()
```

* Next we will filter only the years we are interested in (from 1974 to 1999)

```python
# creating the baseline
baseline_q4 = df_q4['1974':'1999']
baseline_q4
```

* Next we group the dates based on the DAY column and calculating the appropriate mean for TAVG column

```python
baseline_q4_mean = baseline_q4.groupby('DAY').mean()['TAVG']
baseline_q4_mean
```

* Next we procced by calculating the expected average mean temperateure by mapping each day with our baseline mean

```python
maping = lambda x: baseline_q4_mean[x]
df_q4['EXPECTED'] = df_q4['DAY'].map(maping)
```

* Creating a pandas series which will have true value is the average temperature of that day exceeds the expected average temperature by 10 percent. if not exceeds then false for that day

```python
e_V=((df_q4['TAVG']-df_q4['EXPECTED'])/df_q4['EXPECTED'])>.1
```

* Creating a new cloumn for extreme events using our obtained pandas series above

```python
df_q4['EXTREME_EVET']=e_V 
df_q4
```

* Seperating each year data

```python
years_q4=[df_q4.loc[str(i)] for i in range(1955,2021)]
```

* Counting the number of extreme event days in each year

```python
counts=pd.Series([i['EXTREME_EVET'].sum() for i in years_q4])
```

* Filtering not null values in each year for average temperature column

```python
non_null_counts=pd.Series([(len(i.TAVG)-i.TAVG.isnull().sum()) for i in years_q4])
non_null_counts
```

* Making a baseline percentage of chance of extreme event

```python
percent = counts/non_null_counts
ex_by_year=percent
mean_ex_baseline=percent[19:45].mean() 
mean_ex_baseline
```

<b> <span style="font-size: 200%"> <span style='font-family:Helvetica'>Plotting our Results</b> 


```python
# ploting our required graph
plt.figure(figsize=(14,10))
ax = plt.subplot()
colors = ex_by_year
f = lambda x: 'orange' if x > mean_ex_baseline else 'blue'
colors = colors.map(f)
ax.bar([i for i in range(1955,2021)],ex_by_year.values, width=.20, color=colors)
plt.axhline(0,color='grey',xmin=0.046,xmax=0.954)
plt.axhline(mean_ex_baseline,color='grey')
cs=['orange','blue']
leg = plt.legend(['Greater than Baseline mean','Less than Baseline mean'],prop=dict(weight='bold'))
ir=0
for text in leg.get_lines():
    text.set_color(cs[ir])
    text.set_linewidth(2.0)
    ir+=1
plt.xlabel('Year',fontsize=16)
plt.ylabel('Percentage of Heat Event',fontsize=16)
plt.title('Chance of Extreme Heat Events Compared to the Baseline From 1955-2020', 
             fontdict={'fontsize': 20, 'fontweight': 'bold'})
ax.set_ylim(bottom=-0.03);
```

> * Then we will then create a plot showing the ratio of rainfall over rainy days over the years.</br>

<b> <span style="font-size: 200%"> <span style='font-family:Helvetica'>New Dataframe</b> 


* For this section we will create again a new dataframe that is a copy from the dataframe of the first section

```python
df_q5 = df_merged_after.copy()
df_q5
```

* We will change the index of the dataframe and we will put as index the DATE column

```python
df_q5.index=df_q5['DATE']
```

* We create a seperate column for years

```python
df_q5['YEAR']=df_q5['DATE'].dt.year
df_q5
```

* For this analysis, we will use only PRCP and Year columns.
* Therefore, we choose with the following command only these columns.

```python
df_q5=df_q5[['PRCP','YEAR']]
df_q5
```

* Then we groupby the column Year and we calculate the sum of the column PRCP

```python
rain=df_q5.groupby('YEAR').sum()['PRCP']
rain
```

* Next we replace the 0 values with NULL so we can work only with those days which have rainfall

```python
df_q5['PRCP']=df_q5['PRCP'].replace(0,np.nan)
df_q5
```

* Next we group the years and we count the PRCP events per year

```python
rain_days=df_q5.groupby('YEAR').count()['PRCP']
rain_days
```

* Next we find the the ratio of rain for each year

```python
ratio=rain/rain_days 
ratio
```

* We proceed by calculating 10 year rolling mean

```python
q5_10_year_rooing_mean=ratio.rolling(window=10,min_periods=1).mean()
```

```python
# reshaping our ration data to 1D array # we can use this for ploting
ratio.values.reshape(66) 
```

```python
# reshaping our rolling mean data to 1D array # we can use this for ploting
q5_10_year_rooing_mean.values.reshape(66)
```

<b> <span style="font-size: 200%"> <span style='font-family:Helvetica'>Plotting our Results</b> 


```python
plt.figure(figsize=(14,10))
ax = plt.subplot()
plt.bar(ratio.index,ratio.values.reshape(66), width=.20, color='blue')
plt.plot([i for i in range(1955,2021)],q5_10_year_rooing_mean.values.reshape(66),color='steelblue')
plt.axhline(0,color='grey',xmin=0.046,xmax=0.954)
cs2=['grey','steelblue']

leg = plt.legend(['Baseline mean','Ratio'],prop=dict(weight='bold'))
iir=0
for text in leg.get_lines():
    text.set_color(cs2[iir])
    text.set_linewidth(2.0)
    iir+=1
plt.xlabel('Year',fontsize=16)
plt.ylabel('Ratio of Precipitation (mm)',fontsize=16)
plt.title('Ratio of Rainfall Over Rainy Days in Each Year', 
             fontdict={'fontsize': 20, 'fontweight': 'bold'})

ax.set_ylim(bottom=-1)
ax.set(xlabel='Year');
```
