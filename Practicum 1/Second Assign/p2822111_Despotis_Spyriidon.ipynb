> <b> <span style="font-size: 100%"> <span style='font-family:Helvetica'> >The main variables used:</b> <br />
    <br />
> *  <b>Danceability </b> ( float number)<float> Danceability describes how suitable a track is for dancing based on a combination of musical elements including tempo, rhythm stability, beat strength, and overall regularity. A value of 0.0 is least danceable and 1.0 is most danceable. <br />
> *  <b>Energy </b> (number float). Energy is a measure from 0.0 to 1.0 and represents a perceptual measure of intensity and activity. Typically, energetic tracks feel fast, loud, and noisy. For example, death metal has high energy, while a Bach prelude scores low on the scale. Perceptual features contributing to this attribute include dynamic range, perceived loudness, timbre, onset rate, and general entropy.<br />
> *  <b>Key </b> (integer) The key the track is in. Integers map to pitches using standard Pitch Class notation. E.g. 0 = C, 1 = C♯/D♭, 2 = D, and so on. If no key was detected, the value is -1. >= -1 <= 11 <br />
> *  <b>Loudness </b> (number float) The overall loudness of a track in decibels (dB). Loudness values are averaged across the entire track and are useful for comparing relative loudness of tracks. Loudness is the quality of a sound that is the primary psychological correlate of physical strength (amplitude). Values typically range between -60 and 0 db. <br />
> *  <b>Mode </b> (integer) Mode indicates the modality (major or minor) of a track, the type of scale from which its melodic content is derived. Major is represented by 1 and minor is 0.
<br />
> *  <b>Speechiness </b>(number float) Speechiness detects the presence of spoken words in a track. The more exclusively speech-like the recording (e.g. talk show, audio book, poetry), the closer to 1.0 the attribute value. Values above 0.66 describe tracks that are probably made entirely of spoken words. Values between 0.33 and 0.66 describe tracks that may contain both music and speech, either in sections or layered, including such cases as rap music. Values below 0.33 most likely represent music and other non-speech-like tracks.<br />
> *  <b>Acousticness </b>(number float) A confidence measure from 0.0 to 1.0 of whether the track is acoustic. 1.0 represents high confidence the track is acoustic. <br />
> *  <b>Instrumentalness </b> (number float) Predicts whether a track contains no vocals. "Ooh" and "aah" sounds are treated as instrumental in this context. Rap or spoken word tracks are clearly "vocal". The closer the instrumentalness value is to 1.0, the greater likelihood the track contains no vocal content. Values above 0.5 are intended to represent instrumental tracks, but confidence is higher as the value approaches 1.0.<br />
> *  <b>Liveness </b>(number float) Detects the presence of an audience in the recording. Higher liveness values represent an increased probability that the track was performed live. A value above 0.8 provides strong likelihood that the track is live. <br />
> *  <b>Loudness </b>(number float) The overall loudness of a track in decibels (dB). Loudness values are averaged across the entire track and are useful for comparing relative loudness of tracks. Loudness is the quality of a sound that is the primary psychological correlate of physical strength (amplitude). Values typically range between -60 and 0 db. <br />
> *  <b>Valence </b> (number float) A measure from 0.0 to 1.0 describing the musical positiveness conveyed by a track. Tracks with high valence sound more positive (e.g. happy, cheerful, euphoric), while tracks with low valence sound more negative (e.g. sad, depressed, angry).<br />
> *  <b>Tempo </b>(number float) The overall estimated tempo of a track in beats per minute (BPM). In musical terminology, tempo is the speed or pace of a given piece and derives directly from the average beat duration.<br />
> *  <b>Duration_ms </b>(number float) The duration of the track in milliseconds. <br />
> *  <b>Time_signature </b> (number float) An estimated time signature. The time signature (meter) is a notational convention to specify how many beats are in each bar (or measure). The time signature ranges from 3 to 7 indicating time signatures of "3/4", to "7/4".<br />
> *  <b>Explicit</b>(number float) if its explicit content <br />
> *  <b>Year </b> (number float) Year of release <br />

<b> <span style="font-size: 200%"> <span style='font-family:Helvetica'> Checking Multicolinearity </b> 

* Next we check the multicoliniarty assumption for our variables
* Variance inflation factor (VIF) is a measure of the amount of multicollinearity in a set of multiple regression variables.
* From the bellow results we dont significant multicolinireaty 

```python
import scipy.stats as stats
import statsmodels.api as sm
from patsy import dmatrices
from statsmodels.stats.outliers_influence import variance_inflation_factor
y, X = dmatrices("valence ~ danceability + energy + key + loudness + mode + speechiness + acousticness + instrumentalness + liveness + tempo + time_signature", data=df, return_type="dataframe")
vif = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
result = sm.OLS(y, X).fit()
print(vif)
```

* We notice that the multicollinearity is not so strong.

# Tuning the hyper parameters

> * In machine learning, a hyperparameter is a parameter whose value is set before the learning process begins.

* Hyperparameter tuning

```python
from sklearn.model_selection import GridSearchCV
random_forest_vals = {'max_depth': [200], 'n_estimators': [200], 'max_features': ['auto', 'sqrt', 'log2']}
linear_regressor_vals = {'fit_intercept':[True, False], 'copy_X': [True, False], 'n_jobs':[1,2,3,4], 'positive': [True, False]}
decision_tree_regressor_vals = {'criterion': ["squared_error", "friedman_mse", "absolute_error", "poisson"], 'max_depth': [200], 'max_features': ['auto', 'sqrt', 'log2']}
knn_vals = {'n_neighbors': [2,3,5,7,9], 'weights': ['uniform', 'distance'], 'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute', 'auto']}
```

* Parameter Checking for Linear Regressor


```python
lr = LinearRegression()
grid_lr = GridSearchCV(estimator=lr, param_grid=linear_regressor_vals, refit = True, verbose = 0) 
grid_lr.fit(X_train, y_train)
print(grid_lr.best_params_)
```

**Hyperparameter explaination:**<br>
copy_X is True for X will be copied fit_intercept is set True to calculate the intercept for this model.The number of jobs to use for the computation. because it will only provide speedup in case of sufficiently large problems positive option is only supported for dense arrays.

* Parameter Checking for Random Forest Regressor

```python
rfr = RandomForestRegressor()
grid_rfr = GridSearchCV(estimator=rfr, param_grid=random_forest_vals, refit = True, verbose = 0) 
grid_rfr.fit(X_train, y_train)
print(grid_rfr.best_params_)
```

**Hyperparameter explaination:**<br>
max depth is set to true for maximum depth of the tree. If None, then nodes are expanded until all leaves are pure or until all leaves contain less than min_samples_split samples. max features is log2 then max_features=log2(n_features).


* Parameter Checking for KNN Regressor

```python
knn = KNeighborsRegressor()
grid_knn = GridSearchCV(estimator=knn, param_grid=knn_vals, refit = True, verbose = 0) 
grid_knn.fit(X_train, y_train)
print(grid_knn.best_params_)
```

**Hyperparameter explaination:**<br>
algorithm is brute to use the brute force. neighbours are 9 to get the neigbours value of 9, weights is distance to get the weight points by the inverse of their distance. in this case, closer neighbors of a query point will have a greater influence than neighbors which are further away.

*  Parameter Checking for Decision tree

```python
dtr = DecisionTreeRegressor()
grid_dtr = GridSearchCV(estimator=dtr, param_grid=decision_tree_regressor_vals, refit = True, verbose = 0) 
grid_dtr.fit(X_train, y_train)
print(grid_dtr.best_params_)
```

critetion is friedman_mse mean squared error with Friedman’s improvement score for potential splits, max depth 200 is set to true for maximum depth of the tree. If max features is “auto”, then max_features=n_features

# Applying the fine tunned model and evaluting them

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
# Function for model Evaluation
def model_evaluation(y_test, y_pred):
    print("R2 Score:", r2_score(y_test, y_pred))
    print("Root Mean Squarred error:",mean_squared_error(y_test, y_pred, squared=False))
    print("Mean Squarred error:", mean_squared_error(y_test, y_pred))
    print("Mean absolute error:", mean_absolute_error(y_test, y_pred))
    
    return [r2_score(y_test, y_pred), mean_squared_error(y_test, y_pred, squared=False), mean_squared_error(y_test, y_pred), mean_absolute_error(y_test, y_pred)] 
```
